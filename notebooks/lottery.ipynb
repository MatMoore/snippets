# So, you are planning to win the lottery...

*Is it better to buy one ticket for lots of lotteries, or lots of tickets for one lottery?*

**TLDR: it depends on your definition of better.**

## How to win the UK lottery

To win the jackpot, you need all 6 picks to match the numbers drawn, at odds of
**1:13,983,816**.

(Source: https://www.national-lottery.co.uk/games/lotto/game-procedures#int_win)

**NOTE: It is possible to win lesser amounts by matching fewer numbers, but for the purposes of this notebook we will consider any winnings less than the full jackbot to be uninteresting.**

## Comparison of strategies 

```python
%matplotlib inline
from matplotlib import pyplot
from numpy import logspace
```

We will compare the results for up to 1,000,000 Â£1 tickets. At 2 draws per week, this would take some time to split over different lotteries.

```python
draws_per_week = 2
draws_per_year = draws_per_week * 52
years_played = 1E6 / draws_per_year
years_played
```

Call the number of tickets bought n and the probability for a single ticket p:

```python
n = logspace(0, 6)
p = 1 / 13983817.0
```

If we spread our tickets out over different draws, the results are **independent**.

The probability of not winning *any* of the draws is the probability of not winning a single lottery multiplied by the number of lotteries:
$(1-p) n $

Subtracting that from 1 gives us the probability we will win at least 1 draw.

```python
spread_chance = 1 - (1 - p) **n
```

If we put all our tickets into the same draw, then we can make sure all the tickets have different numbers.

Since we only care about results where all numbers on a ticket match, any random numbers will do.

Buying double the amount of tickets gives us double the chances of winning at least once.

```python
single_chance = p * n
```

Plotting both strategies:

```python
pyplot.xkcd()
pyplot.xlabel('Number of tickets')
pyplot.ylabel('Probability of jackpot')
pyplot.plot(n, single_chance, 'r-', n, spread_chance, 'b-')
pyplot.legend(['Same draw = p*n', 'Different draws = 1-(1-p)^n'])
pyplot.show()
```

Buying all your tickets for the same draw has a slight edge, but you would have to spend a lot of money before it gave you a significant advantage.

## Which is better?
Buying all your tickets for the same draw has a slight edge, although you would have to spend a lot of money before it gave you any significant advantage.

However: this graph is only showing the probability of winning at least one jackpot. In the case where you enter only one draw, you will never win more than one jackpot.

TOP TIP: If you are planning to win a million lotteries, you should enter at least a million lotteries.
